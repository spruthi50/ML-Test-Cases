```python
#DATA PREPROCESSING
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,Imputer,StandardScaler
from sklearn.metrics import confusion_matrix
```

```python
#converting categorical data to dummy variable
dataset=pd.read_csv('Churn_Modelling.csv')
y=dataset.iloc[:,-1]
dataset=dataset.drop(columns=['RowNumber','CustomerId','Surname','Exited'])
text=[]
numeric=[]

string=dataset.select_dtypes(include='object').columns

for i in string:
    x=dataset[i].nunique()
    if x<5 and x>0:
        dataset=pd.get_dummies(dataset,columns=[i])
dataset        
```

```python
X=dataset.values
y
```

```python


#encoding categorical data
#labelencoder_X_1=LabelEncoder()
#abelencoder_X_2=LabelEncoder()
#X[:,2]=labelencoder_X_1.fit_transform(X[:, 2])
#onehotencoder=OneHotEncoder(categorical_features=(1))
#X=onehotencoder.fit_transform(X).toarray()


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=0)
X_train
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)

```

```python
#importing the libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initialising ANN
classifier = Sequential()
```

```python
#Adding input layer units tell how many nodes you have to take
#input_dim=no. of independent variables
classifier.add(Dense(units=6,activation='relu',kernel_initializer='uniform',input_dim=13))
```

```python
#Adding hidden layer
classifier.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
```

```python
#as it is output layer so units=1 as we have 1 dependent variable
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
#compiling ANN
#optimizer is the algorithm you want to use to make the ANN more powerful 
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Fitting the ANN to training set
#batch_size is the no. of observation after which you want to change the weight
classifier.fit(X_train,y_train,batch_size=10,nb_epoch=100)
```

```python
y_pred=classifier.predict(X_test)
```

```python
ypred=pd.DataFrame(y_pred)
ypred
```

```python
y_pred=(y_pred>0.5)
```

```python
cm=confusion_matrix(y_test,y_pred)
```
